```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("C:/Users/Admin/Downloads/uber.xlsx")
```

```python
df.info()
```

```python
print(df.describe())
```

```python
revenue= df.groupby('Pickup Location')['Booking Value'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
bars=plt.barh(revenue.index,revenue.values,color='#8b668b')
plt.title('Top 10 locations by revenue')
plt.xticks(rotation=45)
plt.ylabel('Booking value')

plt.show()
```

```python
top_locations = df['Pickup Location'].value_counts().head(10)

plt.figure(figsize=(10,5))
plt.bar(top_locations.index,top_locations.values,color='lightsteelblue')
plt.title('Top 10 Pickup Locations')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()


```

```python
df['Date'] = pd.to_datetime(df['Date'])

daily_bookings = df.groupby('Date').size()

plt.figure(figsize=(10,5))
plt.plot(daily_bookings,color = 'teal')
plt.title('Daily Bookings Trend')
plt.xlabel('Date')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)
plt.show()
```

```python
status_counts = df['Booking Status'].value_counts()

plt.figure(figsize=(11,5))
plt.bar(status_counts.index, status_counts.values,color='brown')
plt.title('Booking Status Distribution')
plt.xlabel('Status')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Booking Value'], bins=20,color='green')
plt.title('Booking Value Distribution')
plt.xlabel('Booking Value')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.countplot(x='Booking Status', data=df,color='#483d8b')
plt.title('Booking Status Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(x=df['Booking Value'],color='#8b475d')
plt.title('Booking Value Boxplot')
plt.show()
```
